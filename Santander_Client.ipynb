## Model Test

```python
model = LogisticRegression()
```

```python
X = train.drop(['TARGET','ID'], axis = 1)
Y = train['TARGET']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state = 1, stratify = Y)
```

### Model Train

```python
model.fit(X_train, Y_train)
```

```python
Y_predic = model.predict(X_test)
```

```python
accuracy_score(Y_test, Y_predic)
```

###### This dataset has an imbalance of the data, in this case another dataset will be created, but balanced

### Applying Nearmiss

```python
nr = NearMiss()
```

```python
X, Y = nr.fit_sample(X, Y)
```

### 2º Data spliting with balanced dataset

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 1, stratify = Y)
```

```python
model2 = LogisticRegression()
model2.fit(X_train, Y_train)
```

```python
Y_predic = model2.predict(X_test)
accuracy_score(Y_test, Y_predic)
```

```python
print(classification_report(Y_test, Y_predic))
```

```python
print(pd.crosstab(Y_test, Y_predic))
```
